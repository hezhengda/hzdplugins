```python
for key, value in dict_data.items():
    
    if key in ['CONTROL', 'SYSTEM', 'ELECTRONS', 'IONS', 'CELL']:
        
        print('&{}'.format(key))
        
        for param, p_v in value.items():
            print('  {}={}'.format(param, p_v))
        
        print('/')

    elif key in ['ATOMIC_SPECIES']:
        
        print('{}'.format(key))
        
        for element, e_v in value.items():
            print('{} {} {}'.format(element, e_v['atomic_weight'], e_v['pseudopotential']))

    elif key in ['K_POINTS']:
        
        kpts = value['kpoints']
        disp = value['displacement']
        print('K_POINTS automatic')
        print('{} {} {} {} {} {}'.format(
            kpts[0], kpts[1], kpts[2],\
            disp[0], disp[1], disp[2]
        ))

    elif key in ['CELL_PARAMETERS']:
        print('CELL_PARAMETERS (angstrom)')
        tmp_str = ''
        for l in value:
            print('{}'.format(listToStr(l, ' ')))

    elif key in ['ATOMIC_POSITIONS']:
        
        print('ATOMIC_POSITIONS (angstrom)')
        
        for l in value:
            print('{}'.format(listToStr(l, ' ')))
    
    else:
        pass
```

```python
a = [1, 2, 3, 4, 5]
def listToStr(l, seperator):
    """
    :param l: the list that we want to print
    :type l: Python list
    
    :param seperator: the seperator that we want to add, e.g. ' ' (space), etc.
    :type seperator: Python string
    """
    
    tmp_str = str(l[0])
    for i in l[1:-1]:
        tmp_str += seperator + str(i)
    tmp_str += seperator + str(l[-1])
    
    return tmp_str
```
